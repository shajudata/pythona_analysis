```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a file and load dataset
my_file = '/content/National_Stock_Exchange_of_India_Ltd.csv'
my_data = pd.read_csv(my_file)
```

```python
# view the first five five rows of the data
my_data.head()
```

```python
# view last five five rows of the data
my_data.tail()
```

```python
# info about the data
my_data.info()
```

```python
# type casting
for i in my_data.select_dtypes(include = 'object').columns[1:]:
    my_data[i] = my_data[i].str.replace(",","")
    my_data[i] = my_data[i].astype(float)
my_data.info()
```

```python
# checking for null values
my_data.isna().sum()
```

```python
# view the top 10 companies where stock prices had a huge change in one year period
my_data.sort_values('365 d % chng',ascending=False).head(10)
```

```python
# view the market volatility for the past one year of all the stocks
dff = my_data.copy()
dff['volatility'] = my_data['52w H'] - my_data['52w L']
market_volatility = dff['volatility'].dropna()
sns.lineplot(data=market_volatility,color='red')
plt.title('Market volatility for one year')
plt.show()
```

```python
#view top stocks where volume traded is higher than 100 lacs
top_vol = my_data[my_data['Volume (lacs)']>100]

# using bar plot view the stocks traded with volume more than 100 in lacs
plt.figure(figsize=(16,9))
sns.barplot(data=my_data,x=top_vol['Symbol'],y=top_vol['Volume (lacs)'],palette='pastel')
plt.xticks(rotation=90)
plt.xlabel('stocks_companies')
plt.ylabel('volume_in_lacs')
plt.title('highest traded volume stocks')
```

```python
plt.figure(figsize=(16,9))
plt.xticks(rotation=90)
sns.barplot(data=my_data,x='Symbol',y='30 d % chng',palette='coolwarm')
plt.title('30 day change in stock prices')
plt.xlabel('companies')
plt.ylabel('change percentage')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.barplot(data=my_data,x='Symbol',y='365 d % chng',palette='dark')
plt.title('Movement of stocks in one year')
plt.xlabel('companies')
plt.ylabel('change percentage')
plt.show()
```

```python
# creating a group for specific sector and analysing data
# Seting the index to Symbol, filter, then reset the indexIT_data = my_data.set_index('Symbol').loc[['INFY', 'TCS', 'WIPRO', 'HCLTECH', 'TECHM']].reset_index()
# Use the loc method to filter the data
IT_data = my_data.loc[my_data['Symbol'].isin(['INFY', 'TCS', 'WIPRO', 'HCLTECH', 'TECHM'])]
IT_table = IT_data.pivot_table(values=['Open', 'High', 'Low', 'LTP', 'Chng', 'Volume (lacs)', '52w H', '52w L', 'Turnover (crs.)', '365 d % chng', '30 d % chng'], index='Symbol').reset_index()
IT_table
```

```python
# analyse the stock growth for one year for the IT sector
plt.figure(figsize=(15,10))

plt.xticks(rotation=90)
sns.barplot(data=IT_table, x='Symbol', y='365 d % chng', hue='30 d % chng', palette='icefire')

plt.show()
```

```python
# to view the correlation in the data
numeric_data = my_data.drop(columns=["Symbol"])
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(16,16))
sns.heatmap(correlation_matrix, annot=True)
```

```python
plt.figure(figsize = (16,7))
plt.xticks(rotation = 90)


sns.barplot(data = my_data.sort_values(by = ['Turnover (crs.)'], ascending = True), x = 'Symbol', y = 'Turnover (crs.)',palette='dark')
plt.xlabel("Companies")
plt.title("Companies by Turnover")
plt.show()
```
